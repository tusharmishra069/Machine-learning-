```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income(usd)')
plt.scatter(df.year,df.income, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.income) #training the model
```

```python
reg.predict([[2020]]) #predicting the income in 2030
```

```python
reg.coef_ #slope
```

```python
reg.intercept_
```
